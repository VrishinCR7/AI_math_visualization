```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-1,1,100)
y = 3*x + 2
weights = np.linspace(-5,5,100)
losses = []
print(len(weights),len(losses))
for w in weights:
  y_pred = w*x
  loss = ((y - y_pred)**2)
  losses.append(loss)
plt.plot(weights,loss)
plt.xlabel("Weights")
plt.ylabel("Loss(MSE)")
plt.title("Weights vs Losses for a linear model")
plt.grid()
plt.show()
```

```python
print(np.mean(losses))
losses = np.matrix(losses)
plt.imshow(losses,aspect="auto",origin="lower")
plt.colorbar(label="Weigths vs Loss/Error (MSE)")
plt.grid()
plt.show()
```

```python
gradient = np.mean(2*(y-y_pred)*(x))
print(gradient)
w = -4
lr = 0.01
steps = []
for i in range(100):
  w = w - lr*gradient
  plt.plot(i,w)
  steps.append(w)
print("The final learned weight :",w)
print(steps)
plt.grid()
plt.show()
```

```python
w1 = np.array([[0.1]])
w2 = np.array([[0.3]])
x = np.array([3])
z2 = np.dot(w2,np.dot(w1,x))
print(z2)
```

```python
def relu(z):
  return max(0,z)
z1 = relu(np.dot(w1,x))
z2 = relu(np.dot(w2,z1))
print(z2)
```

```python
x = np.linspace(-5,5,100)
relu = np.maximum(0,x)
sigmoid = 1/(1+np.exp(-x))
tanh = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
plt.plot(x,relu,label = "Relu activation function")
plt.plot(x,sigmoid,label = "Sigmoid activation function")
plt.plot(x,tanh,label = "Tanh activation function")
plt.grid()
plt.legend()
plt.show()
```

```python
import torch
x = torch.tensor([2])
y = torch.tensor([4])
w = torch.tensor([0.2],requires_grad = True)
y_pred = w*x
loss = (y - y_pred)**2
loss.backward()
print(f"The automatic gradient is {w.grad}")
```

```python
x = np.array([1,2])
y = np.array([3,4])
manhattan = np.sum(np.abs(x-y))
euclidean = np.linalg.norm(x-y)
print(f"The euclidean distance is {euclidean}")
print(f"The manhattan distance is {manhattan}")
```

```python
u = np.array([1,2])
v = np.array([2,3])
w = np.array([3,4])
def cos(a,b):
  return np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
print("u vs v:",cos(u,v))
print("u vs w:",cos(u,w))
plt.quiver(0,0,u[0],u[1],angles = 'xy',scale_units = 'xy',scale = 1,label = "u")
plt.quiver(0,0,v[0],v[1],angles = 'xy',scale_units = 'xy',scale = 1,label = "v")
plt.quiver(0,0,w[0],w[1],angles = 'xy',scale_units = 'xy',scale = 1,label = "w")
plt.legend()
plt.grid()
plt.show()
```
